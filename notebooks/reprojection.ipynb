### Reproject time series tifs

```python
ts_tifs
```

```python
old_folder = 'old' #folder name that has the unprojected time series
folders = next(os.walk(r"C:\Users\mattw\Desktop\projects\glacier_segmentation\time_series\{}".format(old_folder)))[1]
for folder in folders:
    ts_tifs = next(os.walk(r"C:\Users\mattw\Desktop\projects\glacier_segmentation\time_series\{}\{}".format(old_folder,folder)))[2]
    os.makedirs(r"C:\Users\mattw\Desktop\projects\glacier_segmentation\time_series\new\{}".format(folder))
    for tif in ts_tifs:
        #reprojecting each tif in each time series to lat-long
        read.reproject_raster(r"C:\Users\mattw\Desktop\projects\glacier_segmentation\time_series\{old}\{folder}\{tif}".format(old = old_folder,folder = folder,tif = tif),r"C:\Users\mattw\Desktop\projects\glacier_segmentation\time_series\new\{folder}\{tif}".format(folder = folder, tif = tif))
        continue
```
